```python
import os
import numpy as np
import glob
from collections import defaultdict
import xarray as xr
import matplotlib.pyplot as plt
```

```python
#path = "/glade/work/stevenxu/AMOC_models/ACCESS-CM2/r1i1p1f1/msftmz_Omon_ACCESS-CM2_historical_r1i1p1f1_gn_201001-201412.nc"
#data = xr.open_dataset(path, engine="netcdf4")
dir_path = "/glade/work/stevenxu/AMOC_models/msftmz/models/ACCESS-CM2/r1i1p1f1"
all_files = glob.glob(os.path.join(dir_path, "*.nc"))
```

```python
groups = defaultdict(list)
for fp in all_files:
    fname = os.path.basename(fp)
    prefix = fname.rsplit("_", 1)[0]
    print(prefix)
    groups[prefix].append(fp)
```

```python
datasets = {}
for prefix, files in groups.items():
    files = sorted(files)
    print(f"Concatenating {len(files)} files for {prefix}")
    ds = xr.open_mfdataset(
        files,
        combine="by_coords", 
        parallel=True            
    )
    datasets[prefix] = ds
```

```python
data = datasets["msftmz_Omon_ACCESS-CM2_ssp585_r1i1p1f1_gn"]
yearly = data.msftmz.isel(basin=0).max(dim=["lev","lat"]).groupby("time.year").mean(dim="time")
yearly.plot()
```

```python
data.time.values
```

```python
scenarios = list(datasets.items())
n = len(scenarios)
ncols = 3
nrows = int(np.ceil(n / ncols))

fig, axes = plt.subplots(
    nrows=nrows, ncols=ncols,
    figsize=(4*ncols, 3*nrows),
    sharex=False, 
    sharey=False    
)
axes = axes.flatten()

for ax, (scenario, ds) in zip(axes, scenarios):
    # taking monthly max and then averaging over the year
    monthly_max = ds.msftmz.isel(basin=0).max(dim=["lev","lat"])
    yearly = monthly_max.groupby("time.year").mean(dim="time")

    title = scenario.rsplit("_", 3)[1]
    ax.plot(yearly["year"], yearly)
    ax.set_title(title, fontsize=9)
    ax.set_xlabel("Year")
    ax.set_ylabel("Average monthly maximum msftmz")

for ax in axes[n:]:
    ax.set_visible(False)

plt.tight_layout()
plt.show()
```
