```python
import os
import pandas as pd
```

# Read in all txt in sourceID folder, generate a df showing models having which of the variables

```python
source_dir = '/glade/u/home/stevenxu/AMOCproject/scripts/sourceID'
txt_files = [f for f in os.listdir(source_dir) if f.endswith('.txt')]

var_maps = {}

for fname in txt_files:
    path = os.path.join(source_dir, fname)
    var_name = fname.split('.')[0]
    with open(path, 'r') as fh:
        for line in fh:
            model = line.strip().split(' ', 1)[0]
            if var_name not in var_maps:
                var_maps[var_name] = []
            var_maps[var_name].append(model)


all_models = sorted({m for models in var_maps.values() for m in models})
df = pd.DataFrame(
    0,
    index=all_models,
    columns=var_maps.keys(),
)
for var, models in var_maps.items():
    df.loc[models, var] = 1

display(df)
```

## Sorting models with necessary variables

```python
sorted_df = df[
        (df['ocean_meridional_overturning_mass_streamfunction'] == 1) &
        (df['water_flux_into_sea_water'] == 1) &
        (df['surface_downward_heat_flux_in_sea_water'] == 1) &
        (
            ((df['sea_water_potential_temperature'] == 1) & (df['sea_water_salinity'] == 1)) |
            ((df['sea_surface_temperature'] == 1) & (df['sea_surface_salinity'] == 1))
        )
    ]

display(sorted_df)
available_models = sorted_df.index.values
available_models
```

```python
# Write the available models to a text file
output_dir = '/glade/u/home/stevenxu/AMOCproject/output'
output_path = os.path.join(output_dir, 'available_models.txt')
with open(output_path, 'w') as f:
    for model in available_models:
        f.write(model + '\n')

print(f"Wrote {len(available_models)} models to {output_path}")
```
